```
import pandas as pd
import altair as alt

housing = pd.read_csv('https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/housing.csv')
```

```
display(housing.head(4))
```

```
alt.data_transformers.enable("vegafusion")

# Comparing the total square foot of the home to its price
alt.Chart(housing).mark_circle().encode(
    alt.X('sqft_lot', title='Total Square Foot of Lot'),
    alt.Y('price', title='Price')
).interactive()
```

```
# Compare the condition of the home to its grade
alt.Chart(housing).mark_circle().encode(
    alt.X('condition', title='Condition'),
    alt.Y('grade', title='Grade')
)
```

```
display(housing["zipcode"].value_counts())

# Comparing the prices based on where the house is located (zipcode)
alt.Chart(housing).mark_circle().encode(
    alt.X('zipcode', title='Where the house is located', scale=alt.Scale(domain=[98000, 98250])),
    alt.Y('price', title='Price')
)
```

```
display(housing["view"].value_counts())

alt.Chart(housing).mark_circle().encode(
    alt.X('view', title='View'),
    alt.Y('price', title='Price')
)
```

```
under_2mil = housing[housing["price"] <= 2000000]

display(under_2mil["sqft_lot"].value_counts())

alt.Chart(under_2mil).mark_bar().encode(
    alt.X('sqft_lot', title='Total Square Foot of Lot'),
    alt.Y('count()', title='Count')
)

# alt.Chart(under_2mil).mark_circle().encode(
#     alt.X('sqft_lot', title='Total Square Foot of Lot'),
#     alt.Y('price', title='Price')
# )
```
